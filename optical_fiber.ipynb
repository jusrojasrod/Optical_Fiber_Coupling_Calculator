# Documentación Técnica: Acople de Entrada a Fibra Óptica

Este módulo de Python proporciona herramientas de cálculo para el diseño y análisis de sistemas de acople de luz láser hacia fibras ópticas, típicamente monomodo. Permite determinar el diámetro del punto focal, la longitud focal óptima de la lente de acople y la eficiencia teórica del sistema basándose en la superposición modal.


```python
import numpy as np
import math
```

## Fundametos matemáticos

El código se basa en la propagación de haces gaussianos. Las expresiones implementadas son las siguientes:

-  Diámetro del punto focal ($d$):

    \begin{equation*}
        d = \frac{4\lambda f}{\pi D}
    \end{equation*}

Donde $\lambda$ es la longitud de onda, $f$ es la longitud focal y $D$ es el diámetro del haz de entrada.

```python
def diameter_focal_spot(wavelength, diameter_input_beam, focal_length):
    """
    Calculate the diameter of the focal spot given the diameter of the input beam, 
    the focal length, and the wavelength.
    """
    return (4*wavelength*focal_length) / (math.pi*diameter_input_beam)
```

- Longitud focal óptima ($f$): Despejando de la ecuación anterior para igualar el diámetro del punto focal al diámetro del campo modal (MFD) de la fibra:

\begin{equation*}
    f = \frac{\pi D(MFD)}{4\lambda}
\end{equation*}

```python
def focal_length_for_diameter_focal_spot(wavelength, diameter_input_beam, mode_focal_diameter):
    """
    Calculate the focal length given the wavelength, the diameter of the input beam, 
    and the diameter of the focal spot.
    Parameters
    ----------
    wavelength : float
        Wavelength in meters.
    diameter_input_beam : float
        Diameter of the input beam in meters.
    mode_focal_diameter : float
        Diameter of the focal spot in meters. Must be the same as the core diameter 
        of the optical fiber for efficient coupling (MFD).
    Returns
    -------
    float
        Focal length in meters.
    """
    return (mode_focal_diameter * math.pi * diameter_input_beam) / (4*wavelength)
```

- Eficiencia de acople ($\eta$): Calculada a partir de la superposición integral de dos perfiles gaussianos (el haz enfocado y el modo de la fibra):

$$\eta = \left( \frac{2\omega_1\omega_2}{\omega_1^2 + \omega_2^2} \right)^2$$

Donde $\omega_1$ es el radio del haz enfocado y $\omega_2$ es el radio del modo de la fibra ($MFD / 2$).

```python
def fiber_coupling_efficiency(MFD, omega_input_beam, wavelength, focal_length):
    """
    Calculates the optical fiber coupling efficiency.
    Parameters
    ----------
    MFD : float
        Optical fiber diameter in meters.
    omega_input_beam : float
        Radius of the input beam in meters.
    wavelength : float
        Wavelength in meters.
    focal_length : float
        Focal length in meters.
    Returns
    -------
    float
        Optical fiber coupling efficiency.
    """
    omega_1 = (wavelength * focal_length) / (math.pi * omega_input_beam)
    omega_2 = MFD / 2
    eta = (2*omega_1*omega_2/(omega_1**2 + omega_2**2))**2
     
    return eta
```

## Referencia de Funciones

1. `diameter_focal_spot`
Calcula el diámetro del punto focal generado por una lente al enfocar un haz de luz de un diámetro específico.

**Parámetros:**

<div align="center">

| Parámetro | Tipo | Descripción |
| :--- | :--- | :--- |
| `wavelength` | `float` | Longitud de onda del haz en metros. |
| `diameter_input_beam` | `float` | Diámetro del haz colimado de entrada en metros. |
| `focal_length` | `float` | Longitud focal de la lente de acople en metros. |

</div>

**Retorna:**

- `float`: Diámetro del punto focal calculado en metros.

---

2. `focal_length_for_diameter_focal_spot`
Calcula la longitud focal necesaria que debe tener una lente para que el diámetro del punto focal coincida exactamente con el Diámetro del Campo Modal (MFD) de la fibra, maximizando el acople.

**Parámetros:**

<div align="center">

| Parámetro | Tipo | Descripción |
| :--- | :--- | :--- |
| `wavelength` | `float` | Longitud de onda en metros. |
| `diameter_input_beam` | `float` | Diámetro del haz de entrada en metros. |
| `mode_focal_diameter` | `float` | Diámetro del punto focal objetivo en metros (idealmente el MFD de la fibra). |

</div>

**Retorna:**

- `float`: Longitud focal requerida en metros.

---

3. `fiber_coupling_efficiency`
Calcula la eficiencia teórica de acople asumiendo perfiles de intensidad gaussianos perfectos y ausencia de aberraciones esféricas o desalineaciones espaciales/angulares.

**Parámetros:**

<div align="center">

| Parámetro | Tipo | Descripción |
| :--- | :--- | :--- |
| `MFD` | `float` | Diámetro del Campo Modal de la fibra óptica en metros. |
| `omega_input_beam` | `float` | Radio (no diámetro) del haz colimado de entrada en metros. |
| `wavelength` | `float` | Longitud de onda en metros. |
| `focal_length` | `float` | Longitud focal de la lente en metros. |

</div>

**Retorna:**

- `float`: Fracción de la potencia acoplada (valor entre 0 y 1). Multiplicar por 100 para obtener el porcentaje.
